```python
sal=int(input("Enter the salary :"))
if(sal<50000):
  tax=0.05*sal
  print("You taxes are :",tax)
  print("Your final salary is :",sal-tax)
elif(sal>=50000 and sal<=70000):
  tax=0.1*sal
  print("Your taxes are :",tax)
  print("Your final salary is :",sal-tax)
elif(sal>70000 and sal<100000):
  tax=0.12*sal
  print("Your taxes are :",tax)
  print("Your final salary is :",sal-tax)
elif (sal>=100000):
  print("You are promoted")
else :
   pass
```
