```python
#Hypothesis Testing
import numpy as np
from scipy import stats
```

```python
#Loading two groups of data
data1=np.array([28,21,26,29,23])
data2=np.array([21,27,25,28,19])
print(data1.mean())
print(data2.mean())
```

```python
t_stat,p_value =stats.ttest_ind(data1,data2,equal_var=False)
```

```python
print("T-statistic :",t_stat)
print("P-value :",p_value)
```

```python
#set significance level
alpha =0.05
```

```python
#Interpret the results
if p_value < alpha:
  print("Reject the NULL hypothesis.The means are significant")
else:
  print("Failed to reject the NULL hypothesis.The means are different")
```

```python
import pandas as pd
df=pd.read_csv('/content/Vaccine.csv')
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
a=df['Efficiency'].mean()
b=df['Rec_Rate'].mean()
print("Mean of Efficiency :",a)
print("Mean of Recovery rate :",b)
```

```python
Group_A=np.array(df['Efficiency'])
Group_B=np.array(df['Rec_Rate'])
```

```python
t_stat,p_value =stats.ttest_ind(Group_A,Group_B,equal_var=False)
```

```python
Group_A
```

```python
print(t_stat)
print(p_value)
```

```python
plt.plot()
```

```python
alpha=0.05
```

```python
if p_value < alpha:
  print("Reject the NULL hypothesis.The means are significant")
else:
  print("Failed to reject the NULL hypothesis.The means are different")
```

# ANOVA (One way of ANOVA)

```python
dfA =df[df['Vaccine']=='A']['Efficiency']
dfB =df[df['Vaccine']=='B']['Efficiency']
dfA
```

```python
f_stat,p_value=stats.f_oneway(dfA,dfB)
print('F-statistic :',f_stat)
print('p_value :',p_value)
```
