Extracting HTML code of a web page

-Import libraries

-save the url

-Using requests.get() ,access the web page

-Using BeautifulSoup(),access the HTML code

```python
from bs4 import BeautifulSoup
import requests
```

```python
a=('https://www.youtube.com/')
page=requests.get(a)
code=BeautifulSoup(page.text,'html')
print(code)
```

```python
t=code.find('title')
print(t)
```

```python
print(t.text)
```

```python
f=('https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1')
c=requests.get(f)
cc=BeautifulSoup(c.text,'html')
print(cc)
```

```python
t=cc.find('table')
print(t)
```

```python
tt=t.find_all('th')
print(tt)
```

```python
heading=[i.text for i in tt]
print(heading)
```

```python
import pandas as pd
```

```python
mydf2=pd.DataFrame(columns=heading)
print(mydf2)
```
