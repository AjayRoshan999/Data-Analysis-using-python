```python
a=np.array([5,12,18,45]).reshape(2,2)
print(a)
print("____________________")
print(a.max(axis=0))                    #axis=0 =>column
print("____________________")
print(a.max(axis=1))                    #axis=1 =>row
```

```python
a=np.array([5,12,18,45]).reshape(2,2)
b=np.array([7,16,74,36]).reshape(2,2)

print(a)
print("____________________")
print(b)
print("____________________")
print("After vertically stacking arrays")
print(np.vstack((a,b)))
```

```python
a=np.array([5,12,18,45]).reshape(2,2)
b=np.array([7,16,74,36]).reshape(2,2)

print(a)
print("____________________")
print(b)
print("____________________")
print("After horizontally stacking arrays")
print(np.hstack((a,b)))
```

```python
a=np.array([5,12,18,45]).reshape(2,2)
b=np.array([7,16,74,36]).reshape(2,2)

print(a)
print("____________________")
print(b)
print("____________________")
print("After vertically stacking arrays")
print(np.stack((a,b),axis=1))
```

```python
a=np.arange(30).reshape(2,3,5)
print(a)                            #no. of rows become no. of groups
print("after dstack")               #no. of
print(np.dstack(a))
```

```python
#Spltting arrays
a=np.arange(18).reshape(3,6)
print(a)
print("after splitting")
np.hsplit(a,(1,3))
```

# *Statistics*

Linear Algebra and U functions

```python
a=np.array([4,74,64,12,35,15])
b=np.array([2,53,82,24,41,17])
print(np.add(a,b))
```

```python
a=np.array([4,74,64,12,35,15])
b=np.array([2,53,82,24,41,17])
print(np.subtract(a,b))
```

```python
a=np.array([4,74,64,12,35,15])
b=np.array([2,53,82,24,41,17])
print(np.divide(a,b))
```

```python
a=np.array([4,74,64,12,35,15])
b=np.array([2,53,82,24,41,17])
print(np.divmod(a,b))
```

```python
a=np.array([4,74,64,12,35,15])
b=np.array([2,53,82,24,41,17])
print(np.diff(b))
```

```python
a=np.array([4,54,34,21,5,85])
print(np.sort(a))
```

```python
a=np.array([4,74,64,12,35,15])
b=np.array([2,53,82,64,12,77])
print(np.union1d(a,b))
```

```python
a=np.array([4,74,64,12,35,15])
b=np.array([2,53,82,64,12,77])
print(np.intersect1d(a,b))
```

```python
a=np.array([4,74,64,12,35,15])
b=np.array([2,53,82,64,12,77])
print(np.setdiff1d(a,b))
```
