```python
''' There are five subjects psychology,nano-technology,medicine,onchology,anthropology
two students have applied to do masters in five fields with a line plot show in which they can excel
the cutoff of each subject is 70 '''
std1=[67,45,73,71,57]
std2=[43,67,68,53,85]
cutoff=[70,70,70,70,70]
sub=['Psychology','Nano-tech','Medicine','Onchology','Anthropology']
plt.plot(sub,cutoff,label='Cutoff',color='green')
plt.plot(sub,std1,label='Student-1')
plt.plot(sub,std2,label='Student-2')
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
```

```python
'''An India vs Australia match is on live
India has given 298 as target for australia
the runs for every five overs is as follows : 25,51,84,131,160,189,220,250,267,297
Australia have completed 25 overs
score : 15,41,94,110,151
plot overs vs runs for both countries with india as blue line and aus as yellow line linewidth=2
with required labels and legend the marker for india is flag and the marker for aus is plus
take overs from arange function create runs as arrays '''

overs=np.arange(5,51,5)overs=np.arange(5,51,5)
overs2=np.arange(5,26,5)
ind=np.array([25,51,84,131,160,189,220,250,267,297])
aus=np.array([15,41,94,110,151])
plt.title("India vs Australia")
plt.plot(overs,ind,label="India",marker=">",color='b',linewidth=2)
plt.plot(overs2,aus,label="Australia",marker="+",color='y',linewidth=2)
plt.grid()
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.legend()
```
